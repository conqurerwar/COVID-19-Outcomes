```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"C:\Users\hiii\Desktop\CLASS\Python Proj\Project\COVID-19_Outcomes_by_Vaccination_Status_-_Historical.csv")
df.head()
```

```python
print(df.columns.tolist())
```

```python
df_cleaned = df.copy()
rate_ratio_cols = [
    'Unvaccinated Rate', 'Vaccinated Rate', 'Boosted Rate',
    'Crude Vaccinated Ratio', 'Crude Boosted Ratio',
    'Age-Adjusted Unvaccinated Rate', 'Age-Adjusted Vaccinated Rate', 'Age-Adjusted Boosted Rate',
    'Age-Adjusted Vaccinated Ratio', 'Age-Adjusted Boosted Ratio'
]
df_cleaned[rate_ratio_cols] = df_cleaned[rate_ratio_cols].fillna(method='ffill')
num_cols_to_interp = [
    'Population Unvaccinated', 'Population Vaccinated', 'Population Boosted',
    'Outcome Unvaccinated', 'Outcome Vaccinated', 'Outcome Boosted'
]
df_cleaned[num_cols_to_interp] = df_cleaned[num_cols_to_interp].interpolate(method='linear')
df_cleaned.dropna(inplace=True)
print(" Missing values after cleaning:")
print(df_cleaned.isnull().sum().sort_values(ascending=False))
```

```python
df.describe()
```

```python
#  Plot Outcome Rates Over Time by Vaccination Status
df_cleaned['Week End'] = pd.to_datetime(df_cleaned['Week End'])


weekly_avg_rates = df_cleaned.groupby('Week End')[
    ['Unvaccinated Rate', 'Vaccinated Rate', 'Boosted Rate']
].mean().reset_index()


plt.figure(figsize=(14, 6))
sns.lineplot(data=weekly_avg_rates, x='Week End', y='Unvaccinated Rate', label='Unvaccinated Rate')
sns.lineplot(data=weekly_avg_rates, x='Week End', y='Vaccinated Rate', label='Vaccinated Rate')
sns.lineplot(data=weekly_avg_rates, x='Week End', y='Boosted Rate', label='Boosted Rate')
plt.title('COVID-19 Outcome Rates Over Time by Vaccination Status')
plt.xlabel('Date')
plt.ylabel('Average Outcome Rate')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Outcome Rate by Age Group (Bar Plot)
age_group_avg = df_cleaned.groupby('Age Group')[
    ['Unvaccinated Rate', 'Vaccinated Rate', 'Boosted Rate']
].mean().reset_index()
age_group_avg.plot(
    x='Age Group', 
    kind='bar', 
    figsize=(12,6), 
    title='Average Outcome Rate by Age Group and Vaccination Status'
)
plt.ylabel('Average Outcome Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Heatmap: Risk Levels by Age Group & Vaccination Status
heatmap_data = df_cleaned.groupby('Age Group')[
    ['Unvaccinated Rate', 'Vaccinated Rate', 'Boosted Rate']
].mean()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='Reds', fmt='.2f')
plt.title('Average COVID-19 Outcome Rate by Age Group and Vaccination Status')
plt.ylabel('Age Group')
plt.xlabel('Vaccination Status')
plt.tight_layout()
plt.show()
```

```python
# Trends Over Time by Vax Status
weekly_stacked = weekly_avg_rates.set_index('Week End')
plt.figure(figsize=(14, 6))
plt.stackplot(
    weekly_stacked.index,
    weekly_stacked['Unvaccinated Rate'],
    weekly_stacked['Vaccinated Rate'],
    weekly_stacked['Boosted Rate'],
    labels=['Unvaccinated', 'Vaccinated', 'Boosted'],
    colors=['#FF6B6B', '#4ECDC4', '#1A535C'],
    alpha=0.8
)
plt.title('Stacked Area Chart: COVID-19 Outcome Rates by Vaccination Status Over Time')
plt.xlabel('Date')
plt.ylabel('Outcome Rate')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Ststistical Analysis
# Average Death Rate by Vaccination Status
df_deaths = df_cleaned[df_cleaned['Outcome'] == 'Deaths']
df_deaths[['Unvaccinated Rate', 'Vaccinated Rate', 'Boosted Rate']].describe()
```

```python
from scipy.stats import ttest_ind

# Remove NaNs
unvax = df_deaths['Unvaccinated Rate'].dropna()
vax = df_deaths['Vaccinated Rate'].dropna()
t_stat, p_val = ttest_ind(unvax, vax, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Statistically significant difference in death rates between unvaccinated and vaccinated groups.")
else:
    print("No significant difference found.")
```

```python
# Chi-Square Test: Does Age Group affect Outcome Category
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df_cleaned['Age Group'], df_cleaned['Outcome'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square: {chi2:.2f}, p-value: {p:.4f}")
if p < 0.05:
    print("Significant relationship between Age Group and Outcome Type.")
else:
    print("No significant relationship between Age Group and Outcome.")
```
